```python
imageIds = [10,9,35,8,36,16,4,99,1,11]
images = [[] for i in range(10)]
for i in range(10):
    src_img = net.testset[imageIds[i]][0]
    for j in range(10):
        print(classes[i]+'-->'+classes[j])
        if i == j:
            images[i].append(src_img.numpy())
        else:
            adv_images = attack.performCW((src_img[None, ...]),torch.tensor([j]),verbose=False,target=True)
            images[i].append(adv_images.cpu().data.numpy())
```

```python
labels = [[] for i in range(10)]
for i in range(10):
    for j in range(10):
        if i==j:
            t = torch.tensor(images[i][j])
        else:
            t = torch.tensor(images[i][j][0])
        labels[i].append(classes[net.identify(t)])
labels
```

```python
def reduce_string(string,n):
    return (string[:n] + '.') if len(string) > n else string
```

```python
fig, axs = plt.subplots(10,10,figsize=[15,15])
for i in range(10):
    for j in range(10):
        if i==j:
            t = images[i][j]
        else:
            t = images[i][j][0]
            
        #plt.subplot(10,10,(10*j+i)+1)
        #xticks([]), yticks([])
        axs[i,j].imshow(np.transpose(t/2+0.5,(1,2,0)))
        axs[i,j].axis('off')
        axs[i,j].title.set_text(reduce_string(labels[i][j],8))
        
plt.subplots_adjust(wspace=0.4,hspace=0.4)
plt.show()
```

```python
# noise is multiply by 
fig, axs = plt.subplots(10,10,figsize=[10,10])
for i in range(10):
    for j in range(10):
        if i==j:
            t = images[i][j]
            o = np.ones(images[i][j].shape)
        else:
            t = images[i][j][0]
            o = np.ones(images[i][j][0].shape)
        t = (o - 4*np.absolute(t - images[i][i]))   
        axs[i,j].imshow(np.transpose(t,(1,2,0)))
        axs[i,j].axis('off')
        axs[i,j].title.set_text(reduce_string(labels[i][j],8))
        
plt.subplots_adjust(wspace=0.4,hspace=0.4)
plt.show()
```
