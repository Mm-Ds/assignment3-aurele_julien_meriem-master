```python
#Utils
def denormalize(img):
  return img / 2 + 0.5

def imshow(img):
  if img.device!='cpu':
    img = img.cpu().detach()
  npimg =  img.numpy()  

  plt.imshow(np.transpose(npimg, (1, 2, 0)))
  plt.show()
```

```python
def imshow2(img, title):
    npimg = img.numpy()
    fig = plt.figure(figsize = (5, 15))
    plt.imshow(np.transpose(npimg,(1,2,0)))
    plt.title(title)
    plt.show()
```

```python
iterator = iter(net.testloader)
data, label = iterator.next()
```

```python
imshow2(torchvision.utils.make_grid(data.cpu().data, normalize=True), [classes[label[i]] for i in range(1)])
```

## Attacks test 

## FGSM

```python
#FGSM PARAMTERS:
fgsm_params={'eps':0.05}
```

```python
fgsm = FGSM_Attack(net)
```

#### Under attack model evaluation and confusion matrix

```python
accuracy = fgsm.evaluate_model_under_attack(**fgsm_params)
```

```python
confusion_matrix = fgsm.build_confusion_matrix(**fgsm_params)
```

```python
classes = ['plane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
fig, ax = plt.subplots(1,1)
img = ax.imshow(confusion_matrix)

ax.set_xticks(range(10))
ax.set_yticks(range(10))
ax.set_xticklabels(classes,rotation=45)
ax.set_yticklabels(classes)

fig.colorbar(img)
```

### Targeted attack

```python
for i, classe in enumerate(classes):
    new_img= fgsm.generate_adv((img_src[None, ...]).to(net.device),
                               torch.tensor([i]).to(net.device),
                               targeted=True,
                               **fgsm_params)
    new_img2= denormalize(new_img[0])
    imshow(new_img2)
    pred = net.identify(new_img2)
    print('target : '+classe)
    print('pred : ' + classes[pred])
    
plt.show(block=True)
# is there a problem with the imshow function (vs the initial display) or is this display due to some instance-related-variation 
```

### Untargeted attack

```python
new_img= fgsm.generate_adv((img_src[None, ...]).to(net.device),
                                     torch.tensor([label]).to(net.device),
                                     **fgsm_params)
new_img= denormalize(new_img[0])
imshow(new_img)

pred = net.identify(new_img)
print('real : '+classes[label])
print('pred : ' + classes[pred])
    
plt.show(block=True)
```

```python
#Generate adverserial images for each class (for a sample image of a class, generate 10 perturbed ones, one for every target class ) 

cw_params['verbose']=False
imageIds = [10,9,35,8,36,16,4,99,1,11]
images = [[] for i in range(10)]
for i in range(10):
    src_img = net.testset[imageIds[i]][0]
    for j in range(10):
        print(classes[i]+'-->'+classes[j])
        if i == j:
            images[i].append(src_img.numpy())
        else:
            adv_images = cw.generate_adv((src_img[None, ...]),torch.tensor([j]),targeted=True, **cw_params)
            images[i].append(adv_images.cpu().data.numpy())
```

```python
labels = [[] for i in range(10)]
for i in range(10):
    for j in range(10):
        if i==j:
            t = torch.tensor(images[i][j])
        else:
            t = torch.tensor(images[i][j][0])        #N'est ce pas mieux de faire images[i].append(adv_images.cpu().data.numpy())[0]
        labels[i].append(classes[net.identify(t)])   # lors de la consruction de images, ci-dessus ? (suite et impact non encore vus)
labels
```

```python
def reduce_string(string,n):
    return (string[:n] + '.') if len(string) > n else string
```

```python
fig, axs = plt.subplots(10,10,figsize=[15,15])
for i in range(10):
    for j in range(10):
        if i==j:
            t = images[i][j]
        else:
            t = images[i][j][0]
            
        #plt.subplot(10,10,(10*j+i)+1)
        #xticks([]), yticks([])
        axs[i,j].imshow(np.transpose(t/2+0.5,(1,2,0)))
        axs[i,j].axis('off')
        axs[i,j].title.set_text(reduce_string(labels[i][j],8))
        
plt.subplots_adjust(wspace=0.4,hspace=0.4)
plt.show()
```

```python
# noise is multiply by 
fig, axs = plt.subplots(10,10,figsize=[10,10])
for i in range(10):
    for j in range(10):
        if i==j:
            t = images[i][j]
            o = np.ones(images[i][j].shape)
        else:
            t = images[i][j][0]
            o = np.ones(images[i][j][0].shape)
        t = (o - 4*np.absolute(t - images[i][i]))   
        axs[i,j].imshow(np.transpose(t,(1,2,0)))
        axs[i,j].axis('off')
        axs[i,j].title.set_text(reduce_string(labels[i][j],8))
        
plt.subplots_adjust(wspace=0.4,hspace=0.4)
plt.show()
```
