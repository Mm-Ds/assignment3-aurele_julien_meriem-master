```python
from model import Model
from attacks import AttackFGSM
import matplotlib.pyplot as plt
```

```python
#PARAMTERS:
eps = 0.05
```

```python
net = Model()
net.loadModel()
net.loadData(bs=1)
#net.testModel()
```

```python
attack = AttackFGSM(net,eps)
#accuracy = attack.evaluateModelUnderAttack()
```

```python
confusion_matrix = attack.buildConfusionMatrix()
```

```python
classes = ['plane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
fig, ax = plt.subplots(1,1)
img = ax.imshow(confusion_matrix)

ax.set_xticks(range(10))
ax.set_yticks(range(10))
ax.set_xticklabels(classes,rotation=45)
ax.set_yticklabels(classes)

fig.colorbar(img)
```

## Test

```python
classes = net.dataset.classes
classes
```

```python
import numpy as np
index = 3

img_src = net.testset[index][0]
img = net.testset[index][0] / 2 + 0.5     # unnormalize
label = net.testset[index][1]
npimg = img.numpy()
print('True Label :' + classes[label])
plt.imshow(np.transpose(npimg, (1, 2, 0)))
```

```python
pred = net.identify(img)
classes[pred]
```

```python
import torch

for i, classe in enumerate(classes):
    new_img, result = attack.perform((img_src[None, ...]).to(net.device),torch.tensor([i]).to(net.device))
    new_img2 = new_img[0]/ 2 + 0.5
    npimg2 = new_img2.cpu().detach().numpy()
    plt.imshow(np.transpose(npimg2, (1, 2, 0)))
    plt.show()
    pred = net.identify(new_img2)
    print('target : '+classe)
    print('pred : ' + classes[pred])
    
plt.show(block=True)
```

```python
import torch

new_img, result = attack.perform_single((img_src[None, ...]).to(net.device),torch.tensor([label]).to(net.device))
new_img2 = new_img[0]/ 2 + 0.5
npimg2 = new_img2.cpu().detach().numpy()
plt.imshow(np.transpose(npimg2, (1, 2, 0)))
plt.show()
_, predicted = torch.max(result.data, 1)
pred = predicted[0]
print('real : '+classes[label])
print('pred : ' + classes[pred])
    
plt.show(block=True)
```
